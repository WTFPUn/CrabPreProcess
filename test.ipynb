```python
import numpy as np
import pandas as pd
import PIL as pil
import matplotlib.pyplot as plt
```

```python
import os
```

```python
image_name = os.listdir('images')
```

```python
import cv2

b, g, r = cv2.imread('images/' + image_name[0], 1)[:,:,0], cv2.imread('images/' + image_name[0], 1)[:,:,1], cv2.imread('images/' + image_name[0], 1)[:,:,2]


```

```python
# Create histograms
plt.figure(figsize=(10, 4))

plt.subplot(131)
plt.hist(r.ravel(), 256, [0,256])
plt.title('Red')

plt.subplot(132)
plt.hist(g.ravel(), 256, [0,256])
plt.title('Green')

plt.subplot(133)
plt.hist(b.ravel(), 256, [0,256])
plt.title('Blue')

plt.show()
```

```python
# get all image and plot distribution
b_sum = np.array([])
g_sum = np.array([])
r_sum = np.array([])

for i in image_name:
    b, g, r = cv2.imread('images/' + i, 1)[:,:,0], cv2.imread('images/' + i, 1)[:,:,1], cv2.imread('images/' +i, 1)[:,:,2]
    b_sum = np.concatenate((b_sum, b.ravel()))
    g_sum = np.concatenate((g_sum, g.ravel()))
    r_sum = np.concatenate((r_sum, r.ravel()))
    
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(131)
plt.hist(r_sum.ravel(), 256, [0,256])
plt.title('Red')

plt.subplot(132)
plt.hist(g_sum.ravel(), 256, [0,256])
plt.title('Green')

plt.subplot(133)
plt.hist(b_sum.ravel(), 256, [0,256])
plt.title('Blue')
```
