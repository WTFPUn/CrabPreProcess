```python
from datagenerator import DataGenerator, GenerateParams
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
crab = cv2.imread("./crab/crab.png")
no_mesh_cover = cv2.imread("./Box/cover_ero.png")
mesh_cover = cv2.imread("./Box/box_fullmesh.png")
```

```python

params = GenerateParams(rotate_angle=[30, 60, 90], scale=[0.5, 0.8], crab_flip=True, object_flip_h=True, object_flip_v=True)

gen = DataGenerator(crab, no_mesh_cover, mesh_cover)

data = gen.generate(params)
```

```python
img_size = (1200, 900 )
img_color = (0, 0, 0)
obj_per_image = 4
n = 10

img = gen.insert_in_image(img_size, img_color, obj_per_image, n)

```

```python
# subplot with 5 rows and 2 columns
fig, axs = plt.subplots(5, 2, figsize=(10, 20))

for i in range(n//2):
    for j in range(2):
        axs[i, j].imshow(img[i*2+j][0])

plt.show()
```
